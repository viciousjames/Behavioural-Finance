```python
tickers = ["AFK", "FEZ", "GMF", "ILF"]  
stockreturns_array2 = []
for i in tickers:
    stock_array = yf.download(i, start="2016-01-01", end="2022-12-31", interval = '1d',  progress=False)
    stock_array = stock_array[["Adj Close"]].rename(columns={"Adj Close":i}) 
    stockreturns_array2.append(stock_array)  
stockreturns_array2 = pd.concat(stockreturns_array2, axis=1)  
stockreturns_array2
stockreturns_array2 = stockreturns_array2.pct_change()
stockreturns_gross = 1+stockreturns_array2

#Equal weighting portfolio, 0.25 each
stockreturns_gross.iloc[0] = [0.25, 0.25, 0.25, 0.25]
stockreturns_gross.head()
equal_weighted_returns = stockreturns_gross.cumprod()  # cumulative returns
equal_weighted_returns['val'] = equal_weighted_returns.sum(axis=1)
df1 = equal_weighted_returns
df1
```

```python
dep_var = 'Excess_Market_return'
indep_var = 'GDPRet_lag'

gdp_weighted['OOS_Pred_Mean'] = gdp_weighted.expanding(min_periods=20)[dep_var].mean().shift()
gdp_weighted['OOS_Error_Mean'] = gdp_weighted[dep_var] - gdp_weighted['OOS_Pred_Mean']
```

```python
def pred_from_OLS(x, dep_var, indep_var, window):
    """Compute the next period predicted value for univariate
    OLS regression.
    
    Note: 
        The x dataframe must contain dep_var, indep_var
        and the lagged indep_var.

    Args:
        x : The dataframe to use for computation
        dep_var : The dependant variable
        indep_var : The independant variable
        window : The minimum window length

    Returns:
        The predicted value or nan if data shorter than window.
    """
    if x[[dep_var, indep_var]].notnull().all(axis=1).sum() < window:
        return np.nan
    res = smf.ols(dep_var + '~' + indep_var, data=x).fit()
    indep_var_t = x[indep_var].iloc[-1]
    return res.params['Intercept'] + res.params[indep_var] * indep_var_t  # this gives you the expected market returns using your predictor
gdp_weighted['OOS_Pred_OLS_1'] = np.nan
for i in range(len(gdp_weighted)):
    gdp_weighted['OOS_Pred_OLS_1'].iloc[i] = pred_from_OLS(gdp_weighted.iloc[0:i+1], dep_var,
                                                 indep_var, window=20)
gdp_weighted['OOS_Pred_OLS'] = gdp_weighted['OOS_Pred_OLS_1'].shift()
del gdp_weighted['OOS_Pred_OLS_1']  # delete that columns

gdp_weighted['OOS_Error_OLS'] = gdp_weighted[dep_var] - gdp_weighted['OOS_Pred_OLS']
```

```python
1-((gdp_weighted['OOS_Error_OLS'] ** 2).sum()/(gdp_weighted['OOS_Error_Mean'] ** 2).sum())
# to have a good predictor, you must have a positive OOS R2.
```
